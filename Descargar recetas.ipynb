```python
import requests
from bs4 import BeautifulSoup
import json
from googletrans import Translator

# Configura el cliente de traducción
translator = Translator()

# Define la función de traducción
def traducir_al_espanol(text):
    # Traducir el texto de inglés a español
    translation = translator.translate(text, src='en', dest='es')
    return translation.text

# Función para extraer los enlaces a las recetas desde la página de recetas
def extraer_enlaces_recetas(url):
    response = requests.get(url)
    response.raise_for_status()
    soup = BeautifulSoup(response.text, 'html.parser')
    enlaces = []
    # Busca los enlaces a las recetas (ajusta el selector CSS según la página)
    for a_tag in soup.select('a[href*="/recipes/"]'):
        href = a_tag.get('href')
        if href and '/recipes/collection' in href and href.startswith("/"):
            enlaces.append(f"https://www.bbcgoodfood.com{href}")
    return enlaces

def extraer_recetas(url_coleccion):
    # Realiza una solicitud HTTP a la URL de la colección de recetas
    response = requests.get(url_coleccion)
    response.raise_for_status()  # Verifica si la solicitud fue exitosa

    # Analiza el contenido HTML de la respuesta
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Lista para almacenar los enlaces a las recetas individuales
    enlaces_recetas = []
    
    # Busca los enlaces a las recetas individuales (no colecciones)
    for a_tag in soup.select('a[href*="/recipes/"]'):
        href = a_tag.get('href')
        if href and "collection" not in href:  # Excluye enlaces a colecciones
            enlaces_recetas.append(f"https://www.bbcgoodfood.com{href}")
    
    return set(enlaces_recetas)
    
# Función para procesar cada receta
def procesar_receta(url):
    response = requests.get(url)
    response.raise_for_status()
    soup = BeautifulSoup(response.text, 'html.parser')
    recipe_script = soup.find('script', id='__POST_CONTENT__')
    if recipe_script:
        recipe_data = json.loads(recipe_script.string)
        recipe_name = recipe_data['title']
        recipe_name_es = traducir_al_espanol(recipe_name)
        # Extraer ingredientes
        ingredients = []
        ingredients_data = recipe_data['ingredients']
        for ingredient_group in ingredients_data:
            if 'ingredients' in ingredient_group:
                ingredients_list = ingredient_group['ingredients']
                for ingredient in ingredients_list:
                    quantity = ingredient.get('quantityText', '')
                    ingredient_text = ingredient['ingredientText']
                    note = ingredient.get('note', '')
                    ingredient_text_es = traducir_al_espanol(ingredient_text)
                    ingredients.append(f"{quantity} {ingredient_text_es} {note}")

        # Extraer pasos de preparación
        method_steps = recipe_data['methodSteps']
        preparation_steps = []
        for index, step in enumerate(method_steps, start=1):
            step_html = step['content'][0]['data']['value']
            step_soup = BeautifulSoup(step_html, 'html.parser')
            step_text = step_soup.get_text()
            step_text_es = traducir_al_espanol(step_text)
            preparation_steps.append(f"Paso {index}: {step_text_es}")

        # Extraer información nutricional
        nutritional_info = recipe_data['nutritionalInfo']
        translation_dict = {
            "kcal": "calorías",
            "fat": "grasa",
            "carbs": "hidratos de carbono",
            "protein": "proteínas"
        }
        nutrition_info_list = []
        for info in nutritional_info:
            label = info['label']
            value = info['value']
            suffix = info.get('suffix', '')
            prefix = info.get('prefix', '')
            if label in translation_dict:
                spanish_label = translation_dict[label]
                if prefix:
                    prefix = traducir_al_espanol(prefix)
                nutrition_info_list.append(f"{spanish_label} ({label}): {prefix} {value}{suffix}")

        # Imprimir la información obtenida
        print(f"Nombre de la receta: {recipe_name_es}\n")
        print("Ingredientes:")
        for ingredient in ingredients:
            print(f"- {ingredient}")

        print("\nPreparación:")
        for step in preparation_steps:
            print(step)

        print("\nInformación nutricional:")
        for info in nutrition_info_list:
            print(f"- {info}")
    else:
        print(f"No se encontró el script con datos de la receta en la URL: {url}")

# URL de la página de recetas que deseas analizar
url_recetas = "https://www.bbcgoodfood.com/recipes"

# Extrae los enlaces a las recetas
recetas = []
enlaces_recetas = extraer_enlaces_recetas(url_recetas)
for collection in enlaces_recetas:
    enlaces_recetas = extraer_recetas(collection)
    recetas.append(list(enlaces_recetas))

# Procesa cada receta
#for enlace in enlaces_recetas:
#    procesar_receta(enlace)
```
